# Resolver 2 problemas reales

## Problema 1

Escribe un programa en Python que calcule el área de un rectángulo. El programa debe pedir al usuario que ingrese la base y la altura del rectángulo y luego mostrar el área calculada.



## Solucion 1

```python
# Pedir al usuario que ingrese la base del rectángulo
base = float(input("Ingresa la base del rectángulo: "))

# Pedir al usuario que ingrese la altura del rectángulo
altura = float(input("Ingresa la altura del rectángulo: "))

# Calcular el área multiplicando la base por la altura
area = base * altura

# Mostrar el resultado
print("El área del rectángulo es:", area)
```

## Problema 2

Escribe un programa en Python que convierta una temperatura dada en grados Celsius a grados Fahrenheit. El programa debe pedir al usuario que ingrese la temperatura en Celsius y luego mostrar la temperatura equivalente en Fahrenheit.



## Solucion 2

```python
# Pedir al usuario que ingrese la temperatura en Celsius
celsius = float(input("Ingresa la temperatura en grados Celsius: "))

# Convertir Celsius a Fahrenheit usando la fórmula
fahrenheit = (celsius * 9/5) + 32

# Mostrar el resultado
print("La temperatura en grados Fahrenheit es:", fahrenheit)
```
