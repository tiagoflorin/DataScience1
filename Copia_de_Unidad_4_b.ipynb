# Desafio generico

```python
data=pd.read_csv('HRDataset_v14.csv',sep=',')
data.head()
```

```python
# Verificamos si hay nulos primero
data.isnull().sum()
```

# Grafico 1: histograma para ver distribucion del salario

```python
# Orientado a objetos
mpl.style.use('classic')
fig, ax = plt.subplots(figsize=(10,6))
ax.hist(x= data.Salary, bins=None, density=False, orientation='vertical')
ax.set_xlabel('Salario')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma de Salario')
```

```python
# Orientado a estados
mpl.style.use('classic')
plt.figure(figsize=(10,6))
plt.hist(x= data.Salary, bins=None, density=False, orientation='vertical')
plt.xlabel('Salario')
plt.ylabel('Frecuencia')
plt.title('Histograma de Salario')
```

Se logra ver que la distribucion de los datos es asimetrica a derecha con la mayoria de valores en la parte inferior de la distribucion (valores tipicos entre 50000-100000). Se logra ver tambien que hay presencia de valores por encima de 2000000

# Grafico 2: Piechart para ver distribucion de empleados por departamento

```python
data_2=data[['EmpID','Department']].groupby(by='Department').count()
data_2
```

```python
data_2.index
```

```python
fig,ax = plt.subplots(figsize =(10, 6))
ax.pie(data_2.EmpID, labels = data_2.index,autopct='%1.1f%%')
ax.set_title('Piechart de distribucion de departamento')
```

```python
plt.figure(figsize=(10,6))
plt.pie(data_2.EmpID, labels = data_2.index,autopct='%1.1f%%')
plt.title('Piechart de distribucion de departamento')
```
