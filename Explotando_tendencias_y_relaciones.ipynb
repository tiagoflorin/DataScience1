## 1. Medicin de salario mnimo/mximo en 5 pases de Latinoamrica

```python
# Datos de ejemplo (en USD, solo para ilustracin)
data = {
    'Pas': ['Argentina', 'Brasil', 'Chile', 'Colombia', 'Mxico'],
    'Salario Mnimo': [350, 280, 450, 320, 250],
    'Salario Mximo': [2500, 2200, 3000, 2800, 2000]
}
df = pd.DataFrame(data)

# Reestructuramos el DataFrame para que sea ms fcil de graficar con seaborn
df_melted = df.melt(id_vars='Pas', var_name='Tipo de Salario', value_name='Monto (USD)')

# Crear el grfico de barras
plt.figure(figsize=(12, 7))
sns.barplot(x='Pas', y='Monto (USD)', hue='Tipo de Salario', data=df_melted, palette='viridis')

# Aadir ttulos y etiquetas
plt.title('Comparacin de Salario Mnimo y Mximo en 5 Pases de Latinoamrica', fontsize=16)
plt.xlabel('Pas', fontsize=12)
plt.ylabel('Monto (USD)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Tipo de Salario')
plt.tight_layout()

# Mostrar el grfico
plt.show()
```

## 2. Medicin de salario mnimo en 5 pases de Latinoamrica, con una comparacin trimestral

```python
# Datos de ejemplo
data = {
    'Pas': ['Argentina', 'Brasil', 'Chile', 'Colombia', 'Mxico'] * 4,
    'Trimestre': ['T1', 'T1', 'T1', 'T1', 'T1',
                  'T2', 'T2', 'T2', 'T2', 'T2',
                  'T3', 'T3', 'T3', 'T3', 'T3',
                  'T4', 'T4', 'T4', 'T4', 'T4'],
    'Salario Mnimo': [350, 280, 450, 320, 250,
                       355, 282, 455, 325, 252,
                       360, 285, 460, 330, 255,
                       365, 288, 465, 335, 258]
}
df = pd.DataFrame(data)

# Crear el grfico de barras
plt.figure(figsize=(14, 8))
sns.barplot(x='Pas', y='Salario Mnimo', hue='Trimestre', data=df, palette='plasma')

# Aadir ttulos y etiquetas
plt.title('Salario Mnimo Trimestral en 5 Pases de Latinoamrica', fontsize=16)
plt.xlabel('Pas', fontsize=12)
plt.ylabel('Salario Mnimo (USD)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Trimestre')
plt.tight_layout()

# Mostrar el grfico
plt.show()
```

## 3. Evolucin del salario mnimo en 5 pases de Latinoamrica en el ltimo semestre

```python
# Datos de ejemplo
data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio'],
    'Argentina': [350, 352, 355, 358, 360, 365],
    'Brasil': [280, 280, 282, 282, 285, 288],
    'Chile': [450, 450, 455, 455, 460, 465],
    'Colombia': [320, 322, 325, 328, 330, 335],
    'Mxico': [250, 250, 252, 252, 255, 258]
}
df = pd.DataFrame(data)

# Reestructuramos el DataFrame para que sea ms fcil de graficar con seaborn
df_melted = df.melt(id_vars='Mes', var_name='Pas', value_name='Salario Mnimo')

# Aseguramos el orden correcto de los meses
df_melted['Mes'] = pd.Categorical(df_melted['Mes'], categories=['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio'], ordered=True)

# Crear el grfico de lneas
plt.figure(figsize=(12, 7))
sns.lineplot(x='Mes', y='Salario Mnimo', hue='Pas', data=df_melted, marker='o', palette='tab10', linewidth=2.5)

# Aadir ttulos y etiquetas
plt.title('Evolucin del Salario Mnimo en el ltimo Semestre', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Salario Mnimo (USD)', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(title='Pas')
plt.tight_layout()

# Mostrar el grfico
plt.show()
```
