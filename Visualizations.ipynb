# Parte 1: Matplotlib - Los Fundamentos

Matplotlib es la librería "abuela" de la visualización de datos en Python. Es extremadamente potente y personalizable, aunque a veces puede requerir escribir un poco más de código para obtener resultados específicos. La mayoría de las otras librerías de visualización en Python (como Seaborn) se basan en Matplotlib de alguna manera.


```python
#!pip install matplotlib numpy
```

```python
# Importar las librerías necesarias
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## 1.2 Nuestro Primer Gráfico: Un Gráfico de Línea Simple

Vamos a crear el gráfico más básico posible: una línea que conecta una serie de puntos. Usaremos NumPy para generar datos simples.


```python
# Crear datos de ejemplo
x = np.linspace(0, 10, 100) # 100 puntos equidistantes entre 0 y 10
y = np.sin(x)              # Calcular el seno de cada punto x

# Crear el gráfico
plt.plot(x, y)

# Añadir etiquetas y título (¡Siempre importante!)
plt.xlabel("Eje X - Tiempo")
plt.ylabel("Eje Y - Amplitud")
plt.title("Mi Primer Gráfico: Función Seno")

# Mostrar el gráfico
plt.show()
```

## 1.3 La Anatomía de un Gráfico: Figura y Ejes (Axes)

Matplotlib tiene una estructura jerárquica. Lo más importante a entender es la diferencia entre **Figure** (Figura) y **Axes** (Ejes).


* **Figure** (Figura): Es el contenedor de más alto nivel, como
la ventana o la página completa donde todo se dibuja. Puedes pensar en ella como el lienzo en blanco.
* **Axes** (Ejes): Es el área dentro de la figura donde se dibujan los datos (el gráfico en sí, con sus ejes X e Y, etiquetas, etc.). Una figura puede contener uno o más Axes. ¡Ojo! No confundir "Axes" (el objeto de Matplotlib) con "axis" (eje x o y).

Hay dos formas principales de usar Matplotlib:

1. **Interfaz pyplot** (Estilo MATLAB): Como hicimos antes (plt.plot(...), plt.xlabel(...)). Es rápido para gráficos simples, pero puede volverse confuso con gráficos más complejos o múltiples gráficos. Mantiene un estado "global" del gráfico actual.

2. **Interfaz Orientada a Objetos** (OO): Es más explícita y recomendada. Creamos objetos Figure y Axes y luego llamamos a métodos directamente sobre esos objetos.

```python
# Crear datos de ejemplo (los mismos de antes)
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear una Figura y un conjunto de Ejes (Axes)
fig, ax = plt.subplots()

# Usar los métodos del objeto 'ax' para graficar y configurar
ax.plot(x, y)
ax.set_xlabel("Eje X - Tiempo") # Nota: set_xlabel en lugar de xlabel
ax.set_ylabel("Eje Y - Amplitud") # Nota: set_ylabel en lugar de ylabel
ax.set_title("Gráfico Seno (Estilo OO)") # Nota: set_title en lugar de title

# Mostrar el gráfico (plt.show() sigue siendo útil aquí)
plt.show()
```

### ¿Por qué preferir el estilo OO?

Porque es explícito. Siempre sabes sobre qué objeto (ax) estás trabajando. Esto es crucial cuando tienes múltiples gráficos en la misma figura.

* Acostúmbrate a usar fig, ax = plt.subplots()

## 1.4 Tipos de Gráficos Comunes

Matplotlib puede hacer muchos tipos de gráficos. Veamos algunos de los más comunes usando la interfaz OO.

### 1.4.1 Gráfico de Dispersión (Scatter Plot)

Útil para ver la relación entre dos variables numéricas.

```python
# Datos de ejemplo (algo de ruido aleatorio)
x_scatter = np.random.rand(50) * 10 # 50 números aleatorios entre 0 y 10
y_scatter = x_scatter + np.random.randn(50) * 2 # y = x + ruido gaussiano

# Crear figura y ejes
fig, ax = plt.subplots()

# Crear el gráfico de dispersión
ax.scatter(x_scatter, y_scatter)

# Etiquetas y título
ax.set_xlabel("Variable X")
ax.set_ylabel("Variable Y")
ax.set_title("Gráfico de Dispersión")

plt.show()
```

### 1.4.2 Gráfico de Barras (Bar Chart)

Ideal para comparar cantidades entre diferentes categorías.

```python
# Datos de ejemplo
categorias = ['Manzanas', 'Bananas', 'Naranjas', 'Uvas']
cantidades = [25, 40, 30, 35]

# Crear figura y ejes
fig, ax = plt.subplots()

# Crear el gráfico de barras verticales
ax.bar(categorias, cantidades)

# Etiquetas y título
ax.set_xlabel("Fruta")
ax.set_ylabel("Cantidad Vendida")
ax.set_title("Ventas de Frutas")

# Opcional: Rotar etiquetas del eje X si son largas
plt.xticks(rotation=45, ha='right') # Rotar 45 grados, alinear a la derecha

plt.show()

# --- Gráfico de Barras Horizontales ---
fig, ax = plt.subplots()

# Crear el gráfico de barras horizontales (barh)
ax.barh(categorias, cantidades) # ¡Notar barh en lugar de bar!

# Etiquetas y título (ejes invertidos)
ax.set_xlabel("Cantidad Vendida")
ax.set_ylabel("Fruta")
ax.set_title("Ventas de Frutas (Horizontal)")

plt.show()
```

### 1.4.3 Histograma

Muestra la distribución de una única variable numérica, agrupando los datos en "bins" (intervalos).


```python
# Datos de ejemplo (distribución normal)
datos_hist = np.random.randn(1000) # 1000 puntos de una distribución normal estándar

# Crear figura y ejes
fig, ax = plt.subplots()

# Crear el histograma
ax.hist(datos_hist, bins=10, edgecolor='black') # bins controla el número de barras

# Etiquetas y título
ax.set_xlabel("Valor")
ax.set_ylabel("Frecuencia")
ax.set_title("Histograma de Datos Aleatorios Normales")

plt.show()
```

### 1.4.4 Gráfico Circular (Pie Chart)

Muestra proporciones de un todo. ¡Usar con precaución! Son difíciles de interpretar con muchas categorías o categorías con tamaños similares.

```python
# Datos de ejemplo
etiquetas_pie = ['Proteínas', 'Carbohidratos', 'Grasas', 'Vitaminas']
porcentajes = [30, 45, 15, 10]
colores_pie = ['skyblue', 'lightcoral', 'lightgreen', 'gold']
explode = (0, 0.1, 0, 0) # "Explotar" la 2da rebanada (Carbohidratos)

# Crear figura y ejes
fig, ax = plt.subplots()

# Crear el gráfico circular
ax.pie(porcentajes, explode=explode, labels=etiquetas_pie, colors=colores_pie,
       autopct='%1.1f%%', shadow=True, startangle=90)

# Asegurar que el pie sea un círculo (aspecto igual)
ax.axis('equal')

# Título
ax.set_title("Composición Nutricional (Ejemplo)")

plt.show()
```

## 1.5 Personalización de Gráficos

Matplotlib ofrece un control casi total sobre la apariencia de tus gráficos.


### ### 1.5.1 Colores, Estilos de Línea y Marcadores (para `plot`)

```python
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x * 2) * 0.5

fig, ax = plt.subplots()

# Graficar con personalizaciones
ax.plot(x, y1, color='blue', linestyle='-', marker='', label='Seno') # Línea sólida azul
ax.plot(x, y2, color='red', linestyle='--', marker='o', label='Coseno') # Línea discontinua roja con círculos
ax.plot(x, y3, color='#00FF00', linestyle=':', marker='x', label='Seno Modificado') # Línea punteada verde (hex) con cruces

# Añadir leyenda (usa las etiquetas 'label' definidas en plot)
ax.legend()

# Añadir una cuadrícula
ax.grid(True, linestyle='--', alpha=0.6) # Cuadrícula semitransparente

# Establecer límites de los ejes
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)

# Etiquetas y título
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Personalización de Gráficos de Línea")

plt.show()
```

### 1.5.2 Añadir Texto y Anotaciones

Puedes añadir texto en cualquier lugar del gráfico.

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Anotaciones en un Gráfico")

# Añadir texto simple en coordenadas de datos
ax.text(np.pi/2, 1.1, 'Máximo', ha='center') # x, y, texto, alineación horizontal

# Añadir una anotación con flecha
ax.annotate('Mínimo',
            xy=(3 * np.pi / 2, -1), # Punto a señalar (la punta de la flecha)
            xytext=(3 * np.pi / 2, -0.5), # Posición del texto
            arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8), # Propiedades de la flecha
            ha='center')

plt.show()
```

## 1.6 Múltiples Gráficos (Subplots)

A menudo queremos mostrar varios gráficos juntos. La forma más común es usar `plt.subplots()`.

```python
# Crear datos
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
data_hist = np.random.randn(500)
categorias = ['A', 'B', 'C']
valores = [10, 25, 15]

# Crear una figura con una cuadrícula de 2x2 subplots
# figsize controla el tamaño total de la figura en pulgadas (ancho, alto)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# axes ahora es un array de NumPy 2x2 que contiene los objetos Axes
# Accedemos a ellos como axes[fila, columna]

# Subplot 1 (arriba izquierda)
axes[0, 0].plot(x, y_sin, color='blue')
axes[0, 0].set_title('Seno')
axes[0, 0].grid(True)

# Subplot 2 (arriba derecha)
axes[0, 1].plot(x, y_cos, color='red')
axes[0, 1].set_title('Coseno')
axes[0, 1].grid(True)

# Subplot 3 (abajo izquierda)
axes[1, 0].hist(data_hist, bins=20, color='green', alpha=0.7)
axes[1 ,0].set_title('Histograma')

# Subplot 4 (abajo derecha)
axes[1, 1].bar(categorias, valores, color='purple')
axes[1, 1].set_title('Barras')

# Añadir un título general a toda la figura
fig.suptitle('Ejemplo de Múltiples Subplots', fontsize=16)

plt.show()
```

## 1.7 Guardar *Gráficos*

Puedes guardar tus gráficos en archivos de imagen (PNG, JPG, PDF, SVG, etc.).

```python
x = np.linspace(0, 5, 50)
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Gráfico Cuadrático")
ax.set_xlabel("X")
ax.set_ylabel("Y = X^2")

# Guardar la figura ANTES de plt.show() (si lo usas)
# Especifica el nombre del archivo y la extensión deseada
fig.savefig('mi_grafico_cuadratico.png') # Guarda como PNG (bueno para web/docs)
fig.savefig('mi_grafico_cuadratico.pdf') # Guarda como PDF (vectorial, bueno para publicaciones)
fig.savefig('mi_grafico_cuadratico_alta_res.png', dpi=300) # Guarda PNG con mayor resolución (dpi)

# plt.show() # Puedes mostrarlo después si quieres

print("Gráficos guardados como 'mi_grafico_cuadratico.png', '.pdf', y '_alta_res.png'")
```

# Parte 2: Introducción a Seaborn

### Seaborn es otra librería de visualización de datos para Python, construida sobre Matplotlib. Sus principales ventajas son:

1. Estética Mejorada: Viene con estilos y paletas de colores por defecto mucho más atractivos.
2. Interfaz de Alto Nivel: Simplifica la creación de ciertos tipos de gráficos estadísticos complejos (distribuciones, relaciones, datos categóricos).
3. Integración con Pandas DataFrames: Funciona de maravilla con DataFrames de Pandas, haciendo muy fácil graficar columnas de datos.

## 2.1 Instalación e Importación

```python
#!pip install seaborn pandas
```

```python
import seaborn as sns
import pandas as pd # Importamos Pandas también

# Aplicar un estilo por defecto de Seaborn (opcional, pero mejora la apariencia)
sns.set_theme(style="whitegrid") # Prueba otros estilos: "darkgrid", "white", "ticks"
```

## 2.2 Gráficos con Seaborn vs Matplotlib

### 2.2.1 Gráfico de Dispersión (Scatter Plot) con Regresión

```python
# Usaremos los mismos datos de dispersión de antes
x_scatter = np.random.rand(50) * 10
y_scatter = x_scatter + np.random.randn(50) * 2

# Crear el gráfico de dispersión con Seaborn
plt.figure(figsize=(8, 5)) # Podemos seguir usando plt para configurar la figura
sns.scatterplot(x=x_scatter, y=y_scatter, s=100, alpha=0.7) # s=tamaño, alpha=transparencia
plt.title("Scatter Plot con Seaborn")
plt.xlabel("Variable X (Seaborn)")
plt.ylabel("Variable Y (Seaborn)")
plt.show()

# --- Scatter plot con línea de regresión (¡muy fácil!) ---
plt.figure(figsize=(8, 5))
sns.regplot(x=x_scatter, y=y_scatter, scatter_kws={'s': 50, 'alpha': 0.6}, line_kws={'color': 'red'})
plt.title("Scatter Plot con Regresión (Seaborn)")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.show()
```

### 2.2.2 Histograma y Distribución (Histogram / Distplot / Displot)

Seaborn ofrece funciones más potentes para visualizar distribuciones.

```python
# Datos de ejemplo (los mismos del histograma de Matplotlib)
datos_hist = np.random.randn(1000)

# Crear un histograma con una curva de densidad (KDE)
plt.figure(figsize=(8, 5))
sns.histplot(datos_hist, bins=30, kde=True, color='skyblue') # kde=True añade la curva
plt.title("Histograma con KDE (Seaborn)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia / Densidad")
plt.show()
```

```python
# Alternativa: displot (más general, puede crear subplots por categoría)
sns.displot(datos_hist, kind="hist", kde=True, height=5, aspect=1.5, bins=30) # height y aspect controlan tamaño
plt.suptitle("Displot (Histograma con KDE)") # Usar suptitle con displot
plt.show()
```

### 2.2.3 Gráficos Categóricos (Bar Plot, Box Plot, Violin Plot)

Seaborn es muy bueno especialmente con datos categóricos. A menudo es más fácil si los datos están en un DataFrame de Pandas.

```python
# Crear un DataFrame de Pandas de ejemplo
data = {
    'Categoria': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C', 'A', 'B'],
    'Valor': np.random.rand(11) * 100,
    'Grupo': ['G1', 'G1', 'G1', 'G2', 'G2', 'G2', 'G1', 'G1', 'G2', 'G2', 'G1']
}
df = pd.DataFrame(data)

print("DataFrame de ejemplo:")
print(df.head()) # Mostrar las primeras filas

# Bar Plot (Muestra la media por defecto, con intervalo de confianza)
plt.figure(figsize=(8, 5))
sns.barplot(x='Categoria', y='Valor', data=df, palette='viridis', errorbar='sd', hue='Categoria') # ci='sd' muestra desviación estándar
plt.title("Bar Plot (Media por Categoría) con Seaborn")
plt.show()

# Box Plot (Muestra la distribución: mediana, cuartiles, outliers)
plt.figure(figsize=(8, 5))
sns.boxplot(x='Categoria', y='Valor', data=df, palette='pastel', hue='Categoria')
plt.title("Box Plot por Categoría con Seaborn")
plt.show()

# Violin Plot (Combina Box Plot con KDE)
plt.figure(figsize=(8, 5))
sns.violinplot(x='Categoria', y='Valor', data=df, palette='muted', inner='quartile', hue='Categoria') # inner muestra cuartiles
plt.title("Violin Plot por Categoría con Seaborn")
plt.show()

# Bar Plot con Subgrupos (usando 'hue')
plt.figure(figsize=(9, 6))
sns.barplot(x='Categoria', y='Valor', hue='Grupo', data=df, palette='magma')
plt.title("Bar Plot con Subgrupos ('hue') con Seaborn")
plt.legend(title='Grupo')
plt.show()
```

## 2.3 Combinando Seaborn y Matplotlib

```python
# Crear un gráfico con Seaborn
plt.figure(figsize=(8, 5))
sns.histplot(datos_hist, bins=20, kde=True, color='teal')

# Usar Matplotlib para personalizar
plt.title("Gráfico Seaborn Personalizado con Matplotlib", fontsize=16)
plt.xlabel("Mis Valores Personalizados", fontsize=12)
plt.ylabel("Mi Frecuencia Personalizada", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añadir rejilla solo en el eje Y
plt.text(-2, 100, '¡Texto añadido con Matplotlib!', fontsize=10, color='red')

plt.show()
```
