```python
x = np.linspace(0,10,50)
y = 2.5 * x +np.random.normal(size = x.size)
```

```python
y
```

```python
plt.plot(rad,np.sin(rad))
plt.plot(rad,np.cos(rad))
```

```python
plt.scatter(x,y,label='Datos Tirados')
```

```python
A = np.vstack([x,np.ones(len(x))]).T
m,c = np.linalg.lstsq(A,y,rcond=None)[0]
```

```python
regression_lineal = m*x + c
```

```python
plt.plot(x, regression_lineal, "r", label = "Linea De Regresion Lineal")
plt.scatter(x,y,label='Datos Tirados')
```
